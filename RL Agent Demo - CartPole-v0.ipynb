```python
from DQNAgent import Experience, ExperienceReplay, Net, DQNAgent
import numpy as np
from time import sleep
```

```python
import gym
env = gym.make('CartPole-v0')
print(env.action_space)
```

```python
#Change hyperparameters here
agent = DQNAgent(layerSizes=[4, 64, 64, 2],
                epsilon=0.8, eps_decay=0.99997, batch_size=32, discount_fact=0.99, 
                 lr=0.15, num_actions=2, target_update_freq=10000, mem_size = 65000,
                    update_freq = 2)
```

```python
action=0
i_episode = 0

#Arbitary condition, probably could make while(not converged) but this works for now
while agent.epsilon > 0.11:
    i_episode += 1
    state = env.reset()
    score = 0
    agent.loss_val = 0
    for t in range(1000):
#         env.render()
        action = agent.getAction(state)
        n_state, reward, done, info = env.step(action)
        score += reward
        agent.step(state, action, reward, n_state, int(done))
        
        state = n_state
        
        if done:
            print(agent.epsilon)
            print(agent.loss_val/t+1)
            print(n_state)
            print(f"{i_episode} Episode finished after {t+1} timesteps")
            break
env.close()
```

```python
action=0
agent.epsilon = 0
scores = []
for i_episode in range(5):
    state = env.reset()
    score = 0
    
    #env is limited to 200 timesteps by default so range(>200 is fine)
    for t in range(1000):
        env.render()
        action = agent.getAction(state)
        n_state, reward, done, info = env.step(action)
        sleep(0.016) #Limit to 60 FPS
        state = n_state
        score += reward
        if done:
            scores.append(score)
            break
print(f' Mean score = {np.mean(score)}')
env.close()
```
