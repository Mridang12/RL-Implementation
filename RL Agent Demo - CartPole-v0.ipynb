```python
from DQNAgent import Experience, ExperienceReplay, Net, DQNAgent
import numpy as np
from time import sleep
```

```python
from datetime import datetime
import os
dt = datetime.now()
directory = f"CartPole_results/result-{dt.strftime('%m-%d-%Y_%H-%M-%S')}"
os.mkdir(directory)
```

```python
import gym
train_dir = directory+'/'+'training'
os.mkdir(directory+'/'+'training')
env = gym.make('CartPole-v0')
env = gym.wrappers.Monitor(env, train_dir, video_callable=lambda episode_id: episode_id % 200 == 0,force=True)
print(env.action_space)
```

```python
#Change hyperparameters here
agent = DQNAgent(layerSizes=[4, 64, 2],
                epsilon=1, eps_decay=0.9997,min_eps=0.001, batch_size=32, discount_fact=0.99, 
                 lr=0.15, num_actions=2, target_update_freq=10000, mem_size = 65000,
                    update_freq = 2)
```

```python
i_episode = 0
last_n_scores = [0]
n = 10
progress = []
running = []
done = False

while np.mean(last_n_scores) < 195:
    
    i_episode += 1
    state = env.reset()
    score = 0
    agent.loss_val = 0
    running = []
    
    for t in range(1000):
        action = agent.getAction(state)
        n_state, reward, done, info = env.step(action)
        score += reward
        agent.step(state, action, reward, n_state, int(done))
        running.append(action)
        state = n_state
        
        if done:
            print(agent.epsilon)
            print(agent.loss_val/t+1)
            print(n_state)
            print(f"{i_episode} Episode finished after {t+1} timesteps")
            if len(last_n_scores) == n:
                last_n_scores.pop(0)
            last_n_scores.append(score)
            if i_episode % 250 == 0:
                progress.append(running)
            break
progress.append(running)
env.close()
```

```python
play_dir = directory+'/'+'play'
os.mkdir(play_dir)
env = gym.make('CartPole-v0')
env = gym.wrappers.Monitor(env, play_dir, video_callable=lambda episode_id: True,force=True)

action=0
agent.epsilon = 0
scores = []
for i_episode in range(5):
    state = env.reset()
    score = 0
    
    #env is limited to 200 timesteps by default so range(>200 is fine)
    for t in range(1000):
        env.render()
        action = agent.getAction(state)
        n_state, reward, done, info = env.step(action)
        sleep(0.016) #Limit to 60 FPS
        state = n_state
        score += reward
        if done:
            scores.append(score)
            break
print(f' Mean score = {np.mean(score)}')
env.close()
```

```python
##USE THIS CELL TO SAVE TRAINED AGENT

import pickle
from datetime import datetime
dt = datetime.now()

with open(f'{train_dir}/RL-cart-{dt.strftime("%m%d%Y%H%M%S")}.pkl', 'wb') as outp:
    pickle.dump(agent, outp, pickle.HIGHEST_PROTOCOL)

with open(f'{train_dir}/RL-cart-progress{dt.strftime("%m%d%Y%H%M%S")}.pkl', 'wb') as outp:
    pickle.dump(progress, outp, pickle.HIGHEST_PROTOCOL)

```
